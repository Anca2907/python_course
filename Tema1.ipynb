```python
%%writefile tema1/modul_tema1.py

from urllib.request import urlopen
from bs4 import BeautifulSoup
import requests


url = input("Please enter a string:\n")


print(f'You entered {url}')


html_page = urlopen(url)

soup_html_parser = BeautifulSoup(html_page, 'html.parser')

title = soup_html_parser.title.string

print('The title of the URL page is: ', title)

response = requests.get(url)

soup_text = BeautifulSoup(response.text)

metas = soup_text.find_all('meta')

[m.get('content') for m in metas if m.get('name') == 'description']
```
